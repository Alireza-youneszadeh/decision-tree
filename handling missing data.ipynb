```python


# Assuming 'df' is your DataFrame
mf = MissForest()

# Identify columns with string (object) dtype
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute missing values, specifying categorical columns
mod_df_MissForest = mf.fit_transform(df, categorical=categorical_cols)

# Check for remaining missing values
print(pd.DataFrame(mod_df_MissForest).isnull().sum())
```

```python
# prompt: impute the lost data with simple impute  and call the  dataset as mod_df_simpleimpute


# Create the SimpleImputer object
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Impute the missing values in the dataset
mod_df_simpleimpute = imputer.fit_transform(df)

# Convert the imputed data back to a Pandas DataFrame
mod_df_simpleimpute = pd.DataFrame(mod_df_simpleimpute, columns=df.columns)
```

```python
le=LabelEncoder()
mod_df_fillna=mod_df_fillna.apply(le.fit_transform)
mod_df_dropna=mod_df_dropna.apply(le.fit_transform)

mod_df_MissForest=mod_df_MissForest.apply(le.fit_transform)
mod_df_simpleimpute=mod_df_simpleimpute.apply(le.fit_transform)
```

```python
mod_df_fillna.head()
```

```python
x1=mod_df_dropna.loc[:,df.columns!='Disease']
y1=mod_df_dropna.loc[:,'Disease']
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
x2=mod_df_fillna.loc[:,df.columns!='Disease']
y2=mod_df_fillna.loc[:,'Disease']
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=42)
```

```python
x3=mod_df_MissForest.loc[:,df.columns!='Disease']
y3=mod_df_MissForest.loc[:,'Disease']
x3_train,x3_test,y3_train,y3_test=train_test_split(x3,y3,test_size=0.2,random_state=42)
```

```python
x4=mod_df_simpleimpute.loc[:,df.columns!='Disease']
y4=mod_df_simpleimpute.loc[:,'Disease']
x4_train,x4_test,y4_train,y4_test=train_test_split(x4,y4,test_size=0.2,random_state=42)
```

```python
x_train=[x1_train,x2_train,x3_train,x4_train]
x_test=[x1_test,x2_test,x3_test,x4_test]
y_train=[y1_train,y2_train,y3_train,y4_train]
y_test=[y1_test,y2_test,y3_test,y4_test]
```

```python
# prompt: how to apply random forest classifier on below cell

from sklearn.ensemble import RandomForestClassifier

# Define the RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=0)

# Loop over the different datasets
for i in range(len(x_train)):
  # Train the model
  clf.fit(x_train[i], y_train[i])

  # Predict on the test set
  y_pred = clf.predict(x_test[i])

  # Print the accuracy
  print(f"Accuracy for dataset {i+1}: {accuracy_score(y_test[i], y_pred)}")
```
